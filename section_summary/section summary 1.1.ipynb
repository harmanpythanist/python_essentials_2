### Member functions and member variables

```python
import math

print('sin of 30:', math.sin(30))    # SIN OF 30
print('value of pi:', math.pi)    # VALUE OF PI

print('sin of pi:', math.sin(math.pi))   # SIN OF PI

# mod.function(mod.variable)
```

### Importing particular items from module

```python
# import math
from math import sin, cos, tan, pi
# math not defined

# function(data)
print(sin(pi))
print(tan(pi))
print(cos(pi))
```
