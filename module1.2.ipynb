## Math Module

```python
from math import pi, tan, sin, cos, asin, radians, degrees
```

```python
ad = 90
ar = radians(ad)
ad = degrees(ar)
```

```python
print(ad == 90)
```

```python
print(ar == pi/2)
```

```python
print(sin(ar) / cos(ar) == tan(ar))
```

```python
print(asin(sin(ar)) == ar)
```
